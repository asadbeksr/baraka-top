```python
import pandas as pd
import json

# Step 1: Load JSON data
with open('stations.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Step 2: Define a function to filter the required fields for each station
def filter_station_data(station):
    return {
        'id': station['id'],
        'name': station['name'],
        'slug': station['slug'],
        'phone_number': station['phone_number'],
        'region': station['region'],
        'website': station.get('website'),
        'address': station['address'],
        'latitude': station['latitude'],
        'longitude': station['longitude'],
    }

# Step 3: Apply the filter function to each station in the JSON data
filtered_stations = [filter_station_data(station) for station in data['results']]

# Step 4: Convert the filtered data to a DataFrame
df = pd.DataFrame(filtered_stations)

# Step 5: Save the DataFrame to a CSV file
df.to_csv('filtered_stations.csv', index=False, encoding='utf-8')
print("Filtered data saved to filtered_stations.csv")

# Display the DataFrame in the notebook for verification
df.head()
```
